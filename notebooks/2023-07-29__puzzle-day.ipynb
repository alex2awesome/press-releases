```python
import copy
```

```python
def str_diff(a, b):
    a_c = copy.copy(a)
    b_c = copy.copy(b)
    
    output = [] 
    for a_i in a_c:
        if a_i in b_c:
            b_idx = b_c.index(a_i)
            b_c = b_c[:b_idx] + b_c[b_idx + 1:]
        else:
            output.append(a_i)
    return ''.join(output)
```

```python
t = (pd.concat([
    pd.Series(names).to_frame('name'),
    pd.Series(letters).to_frame('letters')
], axis=1)
 .apply(lambda s: s.str.lower().str.replace(' ', ''))
 .assign(n_l_diff=lambda df: df.apply(lambda x: str_diff(x['letters'], x['name']), axis=1)) 
#  .assign(n_l_diff=lambda df: df.apply(lambda x: list(filter(lambda y: y not in x['letters'], x['name'])), axis=1))
#  .assign(l_n_diff=lambda df: df.apply(lambda x: list(filter(lambda y: y not in x['name'], x['letters'])), axis=1)) 
#  .apply(list)
#  .pipe(lambda s: ''.join(pd.Series(list(flatten(s.tolist())))))
)
```
